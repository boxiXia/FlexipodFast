```python
import sys
import gym
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from DDPG import DDPGagent
# from utils import *

env = NormalizedEnv(gym.make("Pendulum-v0"))

# env = gym.make('BipedalWalker-v3')

agent = DDPGagent(env)
noise = OUNoise(env.action_space)
batch_size = 256
rewards = []
avg_rewards = []
```

```python
for episode in range(100):
    state = env.reset()
    noise.reset()
    episode_reward = 0
    
    for step in range(1000):
        action = agent.get_action(state)
        action = noise.get_action(action, step)
        new_state, reward, done, _ = env.step(action) 
        agent.memory.push(state, action, reward, new_state, done)
        
        if len(agent.memory) > batch_size:
            agent.update(batch_size)        
        
        state = new_state
        episode_reward += reward
        if done:
            break

    rewards.append(episode_reward)
    avg_rewards.append(np.mean(rewards[-10:]))
    sys.stdout.write("episode: {}, reward: {}, average _reward: {} \n".format(episode, np.round(episode_reward, decimals=2), np.mean(rewards[-10:])))


plt.plot(rewards)
plt.plot(avg_rewards)
plt.plot()
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.show()
```

```python
import os
# https://discuss.pytorch.org/t/how-to-change-the-default-device-of-gpu-device-ids-0/1041/24
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"   
os.environ["CUDA_VISIBLE_DEVICES"]=f"{1}"

import torch
print(f"current_device:{torch.cuda.current_device()}")
print(f"device_count:{torch.cuda.device_count()}")
device = torch.device("cuda")


import sys
import gym
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ddpg import DDPGagent
from ddpg.utils import OUNoise, NormalizedEnv
from flexipod_env import FlexipodEnv

env = FlexipodEnv(dof = 12)
env = NormalizedEnv(env)

agent = DDPGagent(env)
noise = OUNoise(env.action_space)
batch_size = 256
rewards = []
avg_rewards = []
episode_lens = []
```

```python
for episode in range(20000):
    state = env.reset()
    noise.reset()
    episode_reward = 0
#     env.sock.sendto(env.resume_cmd_b,env.remote_address)
    for step in range(2000):
        action = agent.get_action(state)
        action = noise.get_action(action, step)
        new_state, reward, done, _ = env.step(action) 
        agent.memory.push(state, action, reward, new_state, done)
        
        if len(agent.memory) > batch_size:
            agent.update(batch_size)        
        
        state = new_state
        episode_reward += reward
        if done:
            break
    episode_lens.append(step+1)
    rewards.append(episode_reward)
    avg_rewards.append(np.mean(rewards[-10:]))
    sys.stdout.write(f"episode: {episode}, steps:{step}, reward: {np.round(episode_reward, decimals=2)}, average _reward: {np.mean(rewards[-10:])} \n")

env.send_sock.sendto(env.pause_cmd_b,env.remote_address)

plt.plot(rewards)
plt.plot(avg_rewards)
plt.plot()
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.show()
```
